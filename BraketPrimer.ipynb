# Bra–Ket Notation Primer

In quantum mechanics, **Dirac notation** (also called **bra–ket notation**) is a concise way to describe quantum states and operations.  

---

## Basics

- **Ket**: `|ψ⟩` represents a column vector (a state).  
  Example:  

  $$
  |0⟩ = \begin{bmatrix}1 \\ 0\end{bmatrix}, \quad 
  |1⟩ = \begin{bmatrix}0 \\ 1\end{bmatrix}
  $$

- **Bra**: `⟨ψ|` is the conjugate transpose (row vector) of a ket.  
  Example:  

  $$
  ⟨0| = \begin{bmatrix}1 & 0\end{bmatrix}, \quad 
  ⟨1| = \begin{bmatrix}0 & 1\end{bmatrix}
  $$

---

## Inner Product

- The inner product (overlap between states) is written as:  

  $$
  ⟨φ|ψ⟩
  $$  

  Produces a **scalar** (complex number).  

  Example:  

  $$
  ⟨0|1⟩ = 0, \quad ⟨0|0⟩ = 1
  $$

---

## Outer Product

- The outer product forms an operator:  

  $$
  |ψ⟩⟨φ|
  $$  

  Produces a **matrix**.  

  Example:  

  $$
  |0⟩⟨0| = \begin{bmatrix}1 & 0 \\ 0 & 0\end{bmatrix}
  $$

---

## Operators

- Linear operators (like quantum gates) act on kets:  

  $$
  U|ψ⟩
  $$  

- Measurement probabilities:  

  $$
  P(φ) = |⟨φ|ψ⟩|^2
  $$
